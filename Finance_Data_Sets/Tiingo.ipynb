## Stock Prices

```python
Stickers = pd.read_csv("TickerForTiingo/supported_tickers.csv")
Stickers 
```

**Control of Ticker**

```python
Stickers[Stickers["ticker"] == "AAPL"]
```

**Prices**

```python
import requests
headers = {
    'Content-Type': 'application/json'
}
requestResponse = requests.get(f'https://api.tiingo.com/tiingo/daily/aapl/prices?startDate=2019-01-02&token={os.getenv("Tiingo_key")}', headers=headers)
print(requestResponse.json())
```

```python
data = requestResponse.json()
data
```

```python
AAPL = pd.DataFrame(data)
AAPL
```

```python
AAPL["date"] = pd.DatetimeIndex(AAPL["date"]).strftime("%Y-%m-%d")
AAPL
```

```python
AAPL["date"] = pd.to_datetime(AAPL["date"])
AAPL["date"]
```

```python
AAPL.info()
```

```python
AAPL.isnull().sum()
```

```python
AAPL.iloc[:,1:].describe()
```

```python
AAPL["divCash"].value_counts()
```
