### Alpha Vantage Data Set

[AlphaVantage Website](https://www.alphavantage.co/documentation/)

```python
market = pd.read_csv("Market_Names/physical_currency_list.csv")  # For " &market= " or " &to_currency= " functions 
market
```

```python
exchange_url = 'https://www.alphavantage.co/query?function=CURRENCY_EXCHANGE_RATE&from_currency=BTC&to_currency=CNY&apikey=demo'
exchange_r = requests.get(exchange_url)
exchange_data = exchange_r.json()

pd.DataFrame(exchange_data)
```

```python
# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
BTC_price_url = 'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=CNY&apikey=demo'
BTC_price_r = requests.get(BTC_price_url)
BTC_price = BTC_price_r.json()

BTC_price = pd.DataFrame(BTC_price.get('Time Series (Digital Currency Daily)')).T

BTC_CNY = BTC_price.iloc[:,0:8:2]
BTC_USD = BTC_price.iloc[:,1::2]

BTC_CNY.columns = ["CNY Open","CNY High","CNY Low", "CNY Close"]
CM = market[market["currency code"] == "CNY"]
CM = CM.get("currency name").tolist()
CM = CM[0]

BTC_CNY = BTC_CNY.rename_axis(index = CM)
BTC_USD.columns = ["USD Open","USD High","USD Low","USD Close","USD Market Cap."]

display(BTC_CNY)
display(BTC_USD)
```
