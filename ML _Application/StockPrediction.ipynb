### Features

* Choosing Prediciton Target

```python
ML_data = CitiBank[["Close"]].copy()
ML_data["Date"] = ML_data.index
ML_data
```

```python
ML_data["Days"] = ML_data["Date"].dt.day
ML_data["DayOfWeek"] = ML_data["Date"].dt.dayofweek
ML_data["WeekOfYear"] =  ML_data["Date"].dt.isocalendar().week  # Pandas New Update is a isocalender()
ML_data["Months"] = ML_data["Date"].dt.month
ML_data["Quarter"] = ML_data["Date"].dt.quarter
ML_data["DayOfYear"] = ML_data["Date"].dt.dayofyear
ML_data["Years"] = ML_data["Date"].dt.year
ML_data.head(10)
```

```python
ML_data["WeekOfYear"] = ML_data["WeekOfYear"].astype(int)
```

```python
xgb_model = xgb.XGBRegressor()
xgb_model.fit(X, y)
```

```python
last_30_days_data = ML_data.iloc[-30:]  # Assuming the last 30 rows represent the last 30 days
next_30_days_predictions = xgb_model.predict(last_30_days_data.drop(columns=['Close',"Date"]))
next_30_days_predictions
```

```python
dates_for_next_30_days = pd.date_range(start=ML_data.index[-1], periods=30, freq='D')
predictions_df = pd.DataFrame({'Date': dates_for_next_30_days, 'Predicted_Target': next_30_days_predictions})
predictions_df
```

```python
mse = mean_squared_error(ML_data.iloc[-30:]["Close"], next_30_days_predictions)
print("Mean Squared Error for the next 30 days:", mse)
```
