```python
realtimedata = Data_Sets.Yahoo_Finance("TSLA").real_time_price() 
Go_Line_All = go.Figure()
Go_Line_All.add_trace(go.Line(x = TSLA_Data.index, y= TSLA_Data["Open"], name ="Open", line=dict(dash ="dot")))
Go_Line_All.add_trace(go.Line(x = TSLA_Data.index, y= TSLA_Data["High"], name ="High",line=dict(dash ="dash")))
Go_Line_All.add_trace(go.Line(x = TSLA_Data.index, y= TSLA_Data["Close"], name ="Close"))
Go_Line_All.add_trace(go.Line(x = TSLA_Data.index, y= TSLA_Data["Low"], name ="Low",line=dict(dash ="dashdot")))
Go_Line_All.update_xaxes(title = "Date")
Go_Line_All.update_yaxes(title = "Prices")
Go_Line_All.update_layout(title = f"{realtimedata}",xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

Go_Line_All.show()
```

```python
Tickers = ["TSLA","AAPL","F","C","MCD","BA","IBM","BLK","MSFT","GOOG","GE"]

info = []
industry = []
for i in Tickers:
    tickers = Data_Sets.Yahoo_Finance(i)
    info_get = tickers.info()
    info.append(info_get)
for k in np.arange(0,len(info)):
    all_tickers_info = info[k]
    all_industry_get = all_tickers_info["Information"].get("industry")
    industry.append(all_industry_get)

industry = pd.DataFrame(industry)
industry["Tickers"] = Tickers
industry.columns =["Industry","Tickers"]

labels = industry["Industry"]
values = industry["Industry"].value_counts().to_list()
Go_Pie = go.Figure(data=[go.Pie(labels=labels, values=values)])
Go_Pie.update_layout(title = "Industry Distribution")
Go_Pie.show()
```

```python
Tesla = TSLA_Data
Tesla["MA-50"] = Tesla["Close"].rolling(50).mean()
Tesla["MA-100"] = Tesla["Close"].rolling(100).mean()
Tesla["MA-200"] = Tesla["Close"].rolling(200).mean()

Go_CandleStick_2 = go.Figure(go.Candlestick(x=TSLA_Data.index, open=TSLA_Data["Open"],
                                          low=TSLA_Data["Low"],
                                          high=TSLA_Data["High"],
                                          close=TSLA_Data["Close"], name ="Candle Stick"))
Go_CandleStick_2.add_trace(go.Line(x = Tesla.index, y=Tesla["MA-50"],marker_color="blue",name = "MA-50"))
Go_CandleStick_2.add_trace(go.Line(x = Tesla.index, y=Tesla["MA-100"],marker_color="black",name = "MA-100"))
Go_CandleStick_2.add_trace(go.Line(x = Tesla.index, y=Tesla["MA-200"],marker_color="pink",name = "MA-200"))
Go_CandleStick_2.update_xaxes(title = "Date")
Go_CandleStick_2.update_yaxes(title = "Price")
Go_CandleStick_2.update_layout(title = f"{realtimedata}",xaxis_rangeslider_visible=False, template ="seaborn")
Go_CandleStick_2.show()
```

```python
Same_industry_Companies = ["ADBE","ACN","SONY","UBER","AMD","QCOM"] # Sector: Technology

DataSets = []

for i in Same_industry_Companies:   
    Company_Data = Data_Sets.Yahoo_Finance(i)
    Company_Data_get = Company_Data.get_data()
    DataSets.append(Company_Data_get)

Companies_Data_Sets = {}

for i in np.arange(0,len(DataSets)):
    Companies_Data_Sets[i] = pd.DataFrame(DataSets[i]["Close"])
    
Technology_Companies = pd.DataFrame()
Technology_Companies["Adobe Inc."] = Companies_Data_Sets[0]
Technology_Companies["Accenture plc"] = Companies_Data_Sets[1]
Technology_Companies["Sony Group Corporation"] = Companies_Data_Sets[2]
Technology_Companies["Uber Technologies, Inc."] = Companies_Data_Sets[3]
Technology_Companies["Advanced Micro Devices, Inc."] = Companies_Data_Sets[4]
Technology_Companies["QUALCOMM Incorporated"] = Companies_Data_Sets[5]
Technology_Companies = Technology_Companies[Technology_Companies.index > "2020-01-01"]
Technology_Companies.columns.name = "Company"
Technology_Companies
```

```python
PX_Facet_Plots = px.line(Technology_Companies, facet_col="Company", facet_col_wrap=3)
PX_Facet_Plots.add_hline(y=1, line_dash="dot",
              annotation_text="Jan 1, 2020 baseline",
              annotation_position="bottom right")
PX_Facet_Plots.update_layout(title = "Technology Companies")

PX_Facet_Plots.show()
```
