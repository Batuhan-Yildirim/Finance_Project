```python
Ford_data = DataSets.YahooFinance("TSLA","StockPrices",start_date = "2010-06-29")
Ford_data 
```

```python
Same_industry_Companies = ["ADBE","ACN","SONY","UBER","AMD","QCOM"] # Sector: Technology

DataSets_list = []

for i in Same_industry_Companies:   
    Company_Data = DataSets.YahooFinance(i,"StockPrices")
    DataSets_list.append(Company_Data)

Companies_Data_Sets = {}

for i in np.arange(0,len(DataSets_list)):
    Companies_Data_Sets[i] = pd.DataFrame(DataSets_list[i]["Close"])
    
Technology_Companies = pd.DataFrame()
Technology_Companies["Adobe Inc."] = Companies_Data_Sets[0]
Technology_Companies["Accenture plc"] = Companies_Data_Sets[1]
Technology_Companies["Sony Group Corporation"] = Companies_Data_Sets[2]
Technology_Companies["Uber Technologies, Inc."] = Companies_Data_Sets[3]
Technology_Companies["Advanced Micro Devices, Inc."] = Companies_Data_Sets[4]
Technology_Companies["QUALCOMM Incorporated"] = Companies_Data_Sets[5]
Technology_Companies = Technology_Companies[Technology_Companies.index > "2020-01-01"]
Technology_Companies.columns.name = "Company"
Technology_Companies
```

```python
Tickers = ["TSLA","AAPL","F","C","MCD","BA","IBM","BLK","MSFT","GOOG","GE"]

info = []
industry = []
for i in Tickers:
    tickers = DataSets.YahooFinance(i,"CompanyInfo")
    info.append(tickers)

for k in np.arange(0,len(info)):
    all_tickers_info = info[k]
    all_industry_get = all_tickers_info["Information"].get("industry")
    industry.append(all_industry_get)

industry = pd.DataFrame(industry)
industry["Tickers"] = Tickers
industry.columns =["Industry","Tickers"]
industry
```
